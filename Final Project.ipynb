```python
import pandas as pd 

url = 'EPL.csv'   #Here we import our dataset into pandas
matches_data = pd.read_csv(url)

# Inspect the first 5 rows of the result
matches_data.head(5)
```

```python
matches_data.pivot_table(columns=['Season'], values=['HomeTeam', 'AwayTeam','FTHG','FTAG'],
                         aggfunc={'FTHG':max, 'FTAG':max})

#This pivot table shows the maximum number of home and away goals scored each season
```

```python
matches_data.pivot_table(index='HomeTeam', columns='Season',values='FTHG', aggfunc=sum)

#This pivot table will show us the total number of goals scored by each team at their own Homeground every season
```

```python
matches_data.pivot_table(index='AwayTeam', columns='Season', values='FTHG', aggfunc='sum')

#This pivot table will show us the total number of goals scored by each team away from their Home ground every season.
```

```python
%%read_sql
CREATE DATABASE IF NOT EXISTS EPL DEFAULT CHARACTER SET 'utf8'; 
USE EPL;
```

```python
matches_data.to_sql(
          name = 'matches',              
          schema = 'EPL',   #Here we define the elements in our dataset
          con = engine,                   
          if_exists = 'replace', 
          index = False, 
          dtype = {"Season": String(16),"FTAG":Integer,"FTHG":Integer})
```

```python
%%read_sql
SELECT * 
FROM matches
LIMIT 10;
```

```python
%%read_sql
SELECT HomeTeam, MAX(FTHG)
FROM matches
GROUP BY HomeTeam
Limit 100;

/*This query gives us the maximum number of goals scored in a single game at Home 
by each team in the EPL over the past 25 seasons*/
```

```python
%%read_sql
SELECT HomeTeam, SUM(FTHG),SUM(FTAG)
FROM matches
GROUP BY HomeTeam
Limit 100;

/*This query gives us the total number of goals scored at Home 
as well as the total number of goals scored away from Home by each team in the EPL
over the past 25 seasons*/
```

```python
%%read_sql
SELECT Season, HomeTeam, MAX(FTHG)
FROM matches
GROUP BY Season, HomeTeam
Limit 100;

--This query shows us the maximum number of goals scored by each time at home every season
```
